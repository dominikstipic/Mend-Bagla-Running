```python
import src.scoring as scoring
import src.strava_data as strava
import pandas as pd

data = strava.process()
scores = scoring.process(scoring.WEIGHTS)
```

```python
data[0]["start_date"]
```

```python
def score_timeseries():
    date_score = {"date": [], "score": []}
    for d in data:
        date = d["start_date"]
        score = scoring.activity_effort_score(d, scoring.WEIGHTS)
        date_score["date"].append(date)
        date_score["score"].append(score)
    df = pd.DataFrame.from_dict(date_score)
    return df

df = score_timeseries()
df.plot(x="date", y="score", grid=True, kind="line", marker="o", color="red")
```
